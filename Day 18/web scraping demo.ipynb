```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers = headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
name = []
rating = []
review = []
salaries = []
interviews = []
jobs = []
ctype_and_location = []
for i in company:
    name.append(i.find('h2', class_='companyCardWrapper__companyName').text.strip())
    rating.append(i.find('div', class_='rating_text rating_text--md').text.strip()) 
    review.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())
    action_counts = i.find_all('span', class_='companyCardWrapper__ActionCount')

    salaries.append(action_counts[0].text.strip())
    interviews.append(action_counts[1].text.strip())
    jobs.append(action_counts[2].text.strip())
    ctype_and_location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())

    d={'name':name, 'rating':rating, 'review':review, 'salaries':salaries, 'interviews':interviews, 'jobs':jobs, 'ctype_and_location': ctype_and_location}
    df = pd.DataFrame(d)
    df
```

```python
final = pd.DataFrame()
for j in range(1,350):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'}
    webpage = requests.get(url, headers = headers).text

    soup = BeautifulSoup(webpage, 'lxml')
    
    company = soup.find_all('div', class_='companyCardWrapper')

    name = []
    rating = []
    review = []
    salaries = []
    interviews = []
    jobs = []
    ctype_and_location = []
    for i in company:
        name.append(i.find('h2', class_='companyCardWrapper__companyName').text.strip())
        rating.append(i.find('div', class_='rating_text rating_text--md').text.strip()) 
        review.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())
        action_counts = i.find_all('span', class_='companyCardWrapper__ActionCount')
    
        salaries.append(action_counts[0].text.strip())
        interviews.append(action_counts[1].text.strip())
        jobs.append(action_counts[2].text.strip())
        ctype_and_location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())
    
        d={'name':name, 'rating':rating, 'review':review, 'salaries':salaries, 'interviews':interviews, 'jobs':jobs, 'ctype_and_location': ctype_and_location}
        df = pd.DataFrame(d)
        final=pd.concat([final, df], ignore_index=True)

final
```
