```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://books.toscrape.com/')
```

```python
soup = BeautifulSoup(webpage.text, 'lxml')
```

```python
books = soup.find_all("article", class_="product_pod")
```

```python
data = []

for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    availability = book.find("p", class_="instock availability").text.strip()
    rating = book.p["class"][1]   # Example: "Three"
    
    data.append([title, price, rating, availability])
```

```python
df = pd.DataFrame(data, columns=["Title", "Price", "Rating", "Availability"])
print(df.head())
```

```python
final=pd.DataFrame()
for j in range(1,50):
    url = 'https://books.toscrape.com/catalogue/page-{}.html'.format(j)

    webpage = requests.get(url)
    soup = BeautifulSoup(webpage.text, 'lxml')
    books = soup.find_all("article", class_="product_pod")
    data = []

    for book in books:
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text
        availability = book.find("p", class_="instock availability").text.strip()
        rating = book.p["class"][1]   # Example: "Three"
        
        data.append([title, price, rating, availability])
        df = pd.DataFrame(data)
        
        final=pd.concat([final, df], ignore_index=True)
final
```
