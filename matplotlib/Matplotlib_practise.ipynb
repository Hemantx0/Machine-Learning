```python
plt.pie(count, labels = count.index, autopct= "%1.1f", explode = [0, 0.1, 0.2])  # labels is for naming the area , autopct is for showing numbers in % , explode is for diffrentiate the columns from each other
plt.axis("equal") # It is used in the situation in which pie chart is not in complete cirle or proper shape.
plt.show()
```

```python
df["age"] = [21, 34, 28, 39, 23, 31, 26, 37, 20, 40, 25, 33, 29, 22,]
```

```python
df.head()
```

```python
# Scatter plot

plt.scatter(df["salary"], df["age"], color = "red")
plt.show()
```

```python
sort_age =df.sort_values("age")
```

```python
# Line plot

plt.plot(df["salary"], sort_age["age"], color = "red", marker = "o")
plt.grid()
plt.show()
```

```python
# Bar graph

plt.bar(sort_age["age"], df["salary"], color = "green")
plt.show()
```

```python
#Bivariate analysis -> numerical - categorical
```

```python
hr_sal = df[df["dept"] == 'HR']["salary"]
hr_sal
```

```python
 it_sal = df[df["dept"] == 'IT']["salary"]
 it_sal
```

```python
fin_sal = df[df["dept"] == 'Finance']["salary"]
fin_sal
```

```python
plt.boxplot([hr_sal,it_sal,fin_sal], tick_labels=['HR', 'IT', 'Finance']) # labels = tick_labels -> changed in new version
plt.show()
```

```python
# Pie chart

salary_by_dept = df.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct= "%1.2f")
plt.show()
```

```python
# Bar plot

hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
```

```python
plt.bar(["HR", "IT", "Finance"], [hr_mean, it_mean, fin_mean], color = ["green", "black", "red"])
plt.grid()
plt.show()
```

```python
# Multivariate Analysis: 3 numerical columns
```

```python
df["experience"] = [2, 5, 4, 1, 2, 3, 5, 3, 2, 3, 5, 4, 1, 2]
```

```python
df.head()
```

```python
# Bubble Plot:

plt.scatter(df["age"], df["salary"], s = df['experience']*20, color = "blue")
plt.title("age vs salary vs experience")
plt.xlabel("age")
plt.ylabel("salary")
plt.show()
```

```python
# 2 numerical and 1 cateforical column:
```

```python
plt.scatter(df["age"], df["salary"], c = df["dept"].map({"HR": "yellow", "IT": "blue", "Finance": "orange"}))
plt.title("age vs salary vs department")
plt.xlabel("age")
plt.ylabel("salary")
plt.show()
```

```python
color = {"HR":"yellow", "IT":"blue", "Finance":"orange"}

for dept, color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["age"], df_dept["salary"], c = color, label = dept)

plt.legend()
```

```python
# Object Oriented API
```

```python
fig, axs = plt.subplots (1,3, figsize = (25,5))

# Line plot

axs[0].plot(sort_age["age"], df["salary"], color = "red", marker = "*", linewidth = "2", markersize = 2)
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].set_xlabel("age")
axs[0].set_ylabel("salary")

# histogram:

axs[1].hist(df["salary"], bins = 5, color = "green")
axs[1].set_title("Histogram")
axs[1].set_xlabel("salary")
axs[1].set_ylabel("Frequency")

# Boxplot:

axs[2].boxplot(df["salary"])
axs[2].set_title("Boxplot")
axs[2].set_xlabel("salary")

plt.show()
```

```python
data = {
  'Year': [2020, 2021, 2022, 2023],
  'Sales': [100, 150, 200, 250],
  'Profit': [20, 30, 40, 50],
  'Expenses': [80, 120, 160, 200]
}
```

```python
df2 = pd.DataFrame(data)
```

```python
plt.plot(df2["Year"], df2["Sales"], label = "Sales")
plt.plot(df2["Year"], df2["Profit"], label = "Profit")
plt.plot(df2["Year"], df2["Expenses"], label = "Expenses")

plt.title("Financial Analysis")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()
plt.show()
```
