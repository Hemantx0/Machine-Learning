<h3>Views vs copies</h3>

<big> Views - </big> when we created a view of array the original array get affected <br>
<big> Copies - </big> when copy is created original array don't get affected

```python
arr = np.array([1,2,3,4,5])
view = arr[1:3]
view
```

```python
view[1] = 200
view
```

```python
arr # Here, original array get change
```

```python
#copy
copy = arr[1:3].copy()
copy
```

```python
copy[1] = 100
copy
```

```python
arr # Here, original array get same no changes happen in it 
```

<h1>Transpose of matrix</h1>

```python
arr = np.array([[1,2], [3,4]])
arr
```

```python
print(arr.transpose())
```

```python
# Swapaxes -> swap 2 specific axes in matrix
arr = np.array([[[1,2], [3,4]]])
arr.shape
```

```python
swap = np.swapaxes(arr, 0, 1) # (name of array, 1st axes, 2nd axes)
swap.shape
```

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6], [7,8]])
```

```python
# Vertically stacked
print(np.vstack((a,b)))
```

```python
# Horizontally stacked
print(np.hstack((a,b)))
```

```python
print(np.stack((a,b), axis =  0)) # axis = 0 -> row wise
```

```python
print(np.stack((a,b), axis =  1)) # axis = 1 -> column wise
```
