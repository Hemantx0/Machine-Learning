```python
import numpy as np
```

Vectorization means performing operations on entire arrays at once, instead of using Python for loops.

In simple words:

❌ Loop → element by element ✅ Vectorization → whole array in one shot

NumPy is fast because it uses optimized C code internally, so vectorized operations are much faster than Python loops.

🧠 Why Vectorization is Important?

⚡ Much faster execution

🧼 Cleaner and shorter code

💾 Better memory usage

🔥 Required for ML / Data Science

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])
result = []
for i in range(len(a)):
    result.append(a[i] + b[i])
print(result)
```

This is slow because:

<ul><li>Python loop runs every iteration</li>
<li>Each operation happens separately</li></ul>

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

result = a + b
print(result)
```

NumPy performs addition at low-level (C implementation) → much faster.

```python
# broadcasting and Vectorization
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([10, 20, 30])

print(A + B)
```

```python
# np.vectorize -> convert a regular function to be applied on an array
```

```python
def square(x):
    return x * x


vfun = np.vectorize(square)
```

```python
arr = np.array([2, 4, 5, 6, 7, 9])
arr
```

```python
print(vfun(arr))
```
