<H1>Array Reshaping -> Reshape, Ravel, Flatten</H1>

```python
#Reshape
arr = np.array([1,2,3,4,5,6])
reshaped = arr.reshape(2,3)
print(reshaped)
```

```python
reshape2 = reshaped.reshape(3,2)
reshape2
```

```python
#ravel -> convert 1D array
ravel = reshaped.ravel()
ravel
```

```python
#Flateten -> to 1D array
# -> returns a copy of the array
# It don't change the original array it gives its copy 

flat = reshape2.flatten()
flat
```

```python
flat[0] = 100
```
