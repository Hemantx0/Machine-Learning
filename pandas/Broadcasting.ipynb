## broadcasting means stracting the shape of smaller arrays , so that they match the shape of bigger arrays


```python
image = np.array([[10,25,40,45], [35,75,85,90]])
album = image + 5
print(album)
```

```python
print(np.shape(image)) 
```

Broadcasting works if:

Compare from right to left

Dimensions must be:

Equal, OR

One of them must be 1

Missing dimensions are treated as 1

If none of these satisfied → ❌ ValueError

🧠 Memory Trick

👉 “Right to left, equal or 1 — otherwise done.”
