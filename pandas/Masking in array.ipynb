```python
import numpy as np
```

```python
# Logical and:

arr = np.array([100, 2,3,4,5,6])
```

```python
mask = np.logical_and(arr > 3, arr < 6)
```

```python
mask
```

```python
# Logical or: 

mask = np.logical_or(arr < 4, arr > 5)
print(mask)
```
